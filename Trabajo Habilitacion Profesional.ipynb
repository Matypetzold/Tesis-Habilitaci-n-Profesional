# Adquisición de datos

La adquisición de datos en un estudio de ondas superficiales es un proceso fundamental para determinar las propiedades elásticas del subsuelo, específicamente la velocidad de ondas de corte (Vs). Este parámetro es crucial en estudios de ingeniería y geología, como la caracterización sísmica de sitios, evaluación de riesgos geológicos y determinación de condiciones de cimentación. El procedimiento de adquisición se compone de una serie de etapas bien estructuradas que aseguran la calidad y fiabilidad de los datos obtenidos.

### Selección del Método de Adquisición:

#### La adquisición de ondas superficiales puede realizarse mediante dos métodos principales:

Método de fuente activa: Este enfoque implica la generación controlada de ondas sísmicas a través de fuentes artificiales, como un martillo de percusión o una masa vibratoria. Las fuentes activas son útiles para caracterizar capas someras, ya que permiten captar frecuencias altas que penetran hasta profundidades menores.

Método de campo de ondas pasivo: En este caso, se captan las ondas generadas naturalmente por el ruido ambiental (vehículos, maquinaria, actividad humana, etc.). Los métodos pasivos son más efectivos para caracterizar estructuras a mayor profundidad debido a que captan frecuencias bajas, que penetran más profundamente en el subsuelo.

### Preparación del Sitio y Configuración del Equipamiento

Antes de iniciar la adquisición, es fundamental seleccionar cuidadosamente la ubicación del perfil sísmico y la configuración del equipo. Los sensores (geófonos o acelerómetros) se distribuyen a lo largo de una línea (perfil) a intervalos definidos según la profundidad objetivo y la resolución requerida. Aspectos clave incluyen:

Longitud del arreglo: Dependiendo de la profundidad de investigación, se configura la longitud total del perfil. Arreglos largos (50-100 m) son preferibles para investigar mayores profundidades.

Espaciado de sensores: Un espaciado menor aumenta la resolución cerca de la superficie, mientras que un espaciado mayor permite captar frecuencias bajas para profundidades mayores.

Tipo de sensor: Los geófonos de baja frecuencia (4.5 Hz) son comunes en estudios pasivos y de mayor profundidad, mientras que los geófonos de alta frecuencia (10-14 Hz) se utilizan para investigar capas someras.

### Generación de la Onda (Para Métodos Activos)

En estudios activos, la fuente debe ser lo suficientemente potente como para generar ondas que atraviesen las capas de interés. El proceso de golpeo o impacto se realiza en un extremo de la línea de sensores para maximizar la energía dirigida hacia el perfil. La energía generada por el impacto crea ondas de corte y ondas superficiales (Rayleigh y Love), que son registradas por los sensores en intervalos de tiempo precisos.

### Registro y Verificación de Datos

Una vez activada la fuente o captado el ruido ambiental (en métodos pasivos), los sensores registran la respuesta del terreno en forma de una traza de tiempo. Los datos se registran en una unidad de adquisición que los convierte en señales digitales para su posterior análisis. Cada registro se revisa inmediatamente para verificar la claridad de las ondas de superficie, minimizando ruidos indeseados y asegurando que los datos sean representativos del subsuelo.

Es fundamental realizar verificaciones de calidad de los registros en el sitio, ya que factores como el ruido de fondo elevado, el mal contacto de los sensores con el terreno o configuraciones inapropiadas de la fuente pueden afectar la calidad de los datos. Además, se recomienda realizar repeticiones de los registros (al menos tres golpes o eventos) para obtener datos consistentes y reducir errores.

### Almacenamiento de Datos y Registro de Metadatos

Todos los datos de campo deben almacenarse de manera segura, etiquetados según la ubicación, fecha, y condiciones de adquisición, junto con los metadatos relevantes (tipo de sensor, espaciado, y longitud del arreglo, entre otros). Estos metadatos son esenciales para el procesamiento y la interpretación de los resultados, ya que proporcionan contexto sobre las condiciones de adquisición y la configuración del equipo utilizado.

## Aplicamos una de las herramientas del software llamada "Active FK"

El Active FK de Geopsy se utiliza para analizar y procesar los datos de ondas superficiales en pruebas sísmicas activas, específicamente para obtener la curva de dispersión. Su función principal es:

- **Análisis de la Transformada FK (Frecuencia-Número de Onda)**: Permite transformar los datos de tiempo y espacio captados en campo a un dominio de frecuencia y número de onda. Esto ayuda a identificar la velocidad de fase de las ondas superficiales (como las ondas Rayleigh) en función de la frecuencia.

- **Obtención de la Curva de Dispersión**: A partir del análisis FK, Active FK genera la curva de dispersión, que es esencial para conocer cómo varía la velocidad de fase con la frecuencia. Esta curva se utiliza luego como entrada en el proceso de inversión para estimar el perfil de velocidad del subsuelo.

## Pestaña "Processing" (Active FK)

1. **Source Receiver Distance (Distancia entre fuente y receptor)**: Es la distancia horizontal entre la fuente de ondas sísmicas y los receptores o geófonos. Esta distancia afecta la longitud de onda que se puede analizar y, por lo tanto, la profundidad de investigación.

Escenario: En un ensayo MASW, podrías colocar 24 geófonos con 1 metro de separación entre cada uno, mientras que la fuente se ubica a 5 metros del primer geófono. Esto es adecuado para analizar ondas de baja frecuencia que penetran en el subsuelo, permitiendo explorar capas más profundas.**

2. **Time Range (Rango de Tiempo)**

Delay (Retraso): Es el tiempo desde el inicio de la grabación hasta el inicio del análisis, útil para omitir ruido inicial.
Duration (Duración): Es la longitud del tiempo que se analizará después del retraso, limitando el análisis a las ondas relevantes.

Escenario: Si tus ondas de interés llegan después de un segundo, puedes configurar el Delay en 1 segundo para ignorar el ruido inicial. Luego, establece la Duration en 2 segundos para analizar solo ese segmento, del segundo 1 al segundo 3, y evitar capturar ondas reflejadas u otros ruidos.

3. **FK Power**: Type (Potencia de FK: Tipo)

Conventional (Convencional): Método clásico de transformada de Fourier, adecuado para datos con buen espaciado de receptores y menos ruido.
High Resolution (Alta Resolución): Algoritmo de alta resolución para distinguir señales en entornos ruidosos o con múltiples ondas cercanas.

Escenario: Utiliza "Conventional" en un ensayo MASW en suelos homogéneos, donde las señales son claras y con poco ruido. Emplea "High Resolution" en áreas con mucho ruido o donde necesitas resolver señales superpuestas, como en un entorno urbano ruidoso.

4. **Damping (Amortiguamiento)**: Factor de reducción de amplitud de la señal para mitigar la influencia de frecuencias no deseadas.

Escenario: Si tienes mucho ruido en frecuencias altas, aplica un factor de amortiguamiento de 0.1 o 0.2. Esto reducirá el impacto de estas frecuencias y mejorará la claridad de las ondas de baja frecuencia en el análisis FK.

5. **Oversampling (Sobremuestreo)**: Incremento de puntos de muestreo en el dominio de frecuencia para obtener una mayor resolución, haciendo el análisis FK más detallado.

Escenario: Si tienes una señal de 100 Hz y deseas más detalle en la curva de dispersión, usa un sobremuestreo de 2. Esto duplicará los puntos de frecuencia, mejorando la precisión en la identificación de velocidades de fase.

6. **Band Width (Ancho de Banda)**: Rango de frecuencias a analizar en el espectro FK, permitiendo enfocarse solo en las frecuencias relevantes.

Escenario: Si quieres analizar frecuencias entre 5 y 20 Hz, configura el ancho de banda en este rango para evitar analizar ruido en frecuencias fuera de interés y concentrarte solo en las ondas que proporcionan información útil para la interpretación geofísica.

## Pestaña "Output" (Active FK)

1. **Frequency Sampling (Muestreo de Frecuencia)**: Define el rango de frecuencias que se analizará y la manera en que se dividirán estas frecuencias en el análisis.

From (Desde) y To (Hasta): Establecen el límite inferior y superior de frecuencias, en Hz, para el análisis FK.
Step (Paso): Permite seleccionar el tipo de incremento de frecuencia entre "log" (logarítmico), "linear" (lineal) e "inverse" (inverso), lo cual afecta la resolución de frecuencia en el espectro FK.
Escenario: Si deseas analizar un rango de frecuencias entre 5 Hz y 20 Hz con más detalle en las frecuencias bajas, puedes establecer From en 5 Hz, To en 20 Hz y seleccionar un Step logarítmico para obtener mayor resolución en las frecuencias bajas y menos en las altas.

2. **Velocity Sampling (Muestreo de Velocidad)**: Define el rango de velocidades y la manera en que se dividirán en el análisis FK.

From (Desde) y To (Hasta): Establecen el rango de velocidades en m/s que se analizarán en el dominio FK.
Step (Paso): Permite seleccionar el tipo de incremento de velocidad entre "log" (logarítmico), "linear" (lineal) e "inverse" (inverso), lo cual determina la resolución en la distribución de velocidades en el análisis.
Escenario: Si esperas encontrar velocidades entre 100 m/s y 1000 m/s, pero deseas tener más detalle en las velocidades bajas, establece From en 100 m/s, To en 1000 m/s, y selecciona un Step logarítmico. Esto es útil para estudios de ondas superficiales, donde los materiales más cercanos a la superficie suelen tener velocidades más bajas.

3. **Beam Power Normalization (Normalización de Potencia de Haz)**: Controla cómo se normaliza la potencia del haz (beam power) en el análisis FK, afectando la representación de la energía en el espectro.

Raw Beam Power: Muestra la potencia sin normalización, útil para observar la energía absoluta en el espectro.
By Spectrum Power: Normaliza la potencia con respecto a la potencia del espectro de frecuencia, destacando áreas con concentraciones de energía específicas.
By Maximum Beam Power: Normaliza la potencia en función de la potencia máxima del haz, resaltando el punto de mayor energía en el espectro FK.
Escenario: Si buscas resaltar los puntos de mayor energía en el espectro FK para identificar la velocidad de fase dominante, selecciona By Maximum Beam Power. En cambio, si quieres ver la distribución de energía de forma más absoluta, usa Raw Beam Power.

## Obtener archivos ".param"

Los archivos .param en el contexto de inversiones sísmicas, como las realizadas en Geopsy o Dinver, son archivos de configuración que definen los parámetros del modelo de inversión. Estos archivos especifican los ajustes iniciales y límites del modelo de subsuelo que se utilizará en la inversión.

Generalmente, un archivo .param incluye:

- **Número de capas y espesores**: Define el número de capas en el modelo y los límites de cada espesor.
- **Velocidad de onda de corte (Vs)**: Establece rangos para la velocidad de las ondas de corte en cada capa.
- **Densidad y velocidad de onda P (Vp)**: Parámetros opcionales o fijos para cada capa.
- **Restricciones**: Límites mínimos y máximos para cada parámetro, como Vs y espesor, para guiar la inversión.

## Variables:

- "wmin y wmax": Son los valores de longitud de onda mínima y máxima, respectivamente. Estos valores definen el rango de longitudes de onda (en metros) que se utilizarán para determinar las capas en los modelos de velocidad. Aquí, wmin es 2 m y wmax es 20 m.

**vp (Parámetro de Velocidad de Onda P):**

- swprepost.Parameter.from_ln: Es un método que crea un parámetro con una parametrización de “Layering by Number” (por número de capas, abreviado LN).
- wmin=wmin y wmax=wmax: Definen el rango de longitudes de onda.
- nlayers=3: Define el número de capas en el modelo para este parámetro (aquí, 3 capas).
- par_min=200 y par_max=600: Establecen los límites inferior y superior para la velocidad de onda P (Vp) en cada capa, en m/s.
- par_rev=False: Indica que el parámetro no se invierte; en este caso, los valores de velocidad aumentarán o disminuirán según el modelo sin revertirse.
- depth_factor=2: Factor de profundidad utilizado para calcular el grosor de las capas en relación con la longitud de onda mínima, generalmente para asegurar que las capas capturen las variaciones en el subsuelo hasta una profundidad adecuada.

**pr (Razón de Poisson):**

- Este parámetro es similar al anterior, pero con nlayers=1, indicando que la razón de Poisson se considera constante en todo el modelo.
- par_min=0.2 y par_max=0.5: Rango de valores para la razón de Poisson en el modelo (sin unidades).
- Los otros argumentos funcionan de forma similar a los de vp.

**vs (Parámetro de Velocidad de Onda S):**

- swprepost.Parameter.from_lr: Método que crea un parámetro con una parametrización de "Layering Ratio" (proporción de capas, abreviado LR).
- lr=2.0: Define el ratio de aumento en el grosor de las capas, en función de la longitud de onda mínima.
- par_min=100 y par_max=350: Límite inferior y superior para la velocidad de onda S (Vs) en cada capa, en m/s.
- Los otros parámetros funcionan como en vp.

**rh (Densidad):**

- swprepost.Parameter.from_fx(2000): Define un parámetro fijo para la densidad (rho) de 2000 kg/m³ para todas las capas.

**param (Parametrización de Modelo Completa):**

- swprepost.Parameterization(vp=vp, pr=pr, vs=vs, rh=rh): Combina todos los parámetros definidos (vp, pr, vs, rh) en una parametrización completa que representa el modelo de subsuelo, asignando valores para las velocidades y propiedades elásticas de cada capa.

```python
# Create an example parameterization
wmin, wmax = 2, 20
vp = swprepost.Parameter.from_ln(wmin=wmin, wmax=wmax, nlayers=3, par_min=200, par_max=600, par_rev=False, depth_factor=2)
pr = swprepost.Parameter.from_ln(wmin=wmin, wmax=wmax, nlayers=1, par_min=0.2, par_max=0.5, par_rev=False, depth_factor=2)
vs = swprepost.Parameter.from_lr(wmin=wmin, wmax=wmax, lr=2.0, par_min=100, par_max=350, par_rev=False, depth_factor=2)
rh = swprepost.Parameter.from_fx(2000)
param = swprepost.Parameterization(vp=vp, pr=pr, vs=vs, rh=rh)

# Write parameterization to .param format
param.to_param(fname_prefix="to_param_v2", version="2.10.1")   # Write param using v2 style
param.to_param(fname_prefix="to_param_v3", version="3.4.2")   # Write param using v3 style
```

## Obtener carpeta "3_text"

La carpeta 3_text en el flujo de trabajo de inversiones sísmicas, como en el software SWinvert o Geopsy, es un directorio de salida que almacena los resultados de la inversión en formato de texto. Esta carpeta suele contener archivos generados que describen el modelo final del subsuelo obtenido tras el proceso de inversión.

En particular, puede incluir:

- **Modelos de subsuelo**: Archivos .txt que muestran los perfiles de velocidad de onda de corte (Vs) con la profundidad, detallando las propiedades de cada capa calculada.
- **Curvas de dispersión teóricas**: Archivos con las curvas de dispersión obtenidas del modelo, comparables con los datos experimentales.
- **Curvas de elipticidad**: Si se calcularon, pueden aparecer aquí las curvas de elipticidad de las ondas Rayleigh.

### Parametros:

- **ntrial**: Este parámetro representa el número de pruebas o iteraciones independientes realizadas en una inversión. Un valor típico podría ser 10, lo que significa que se ejecutarán diez inversiones para evaluar la variabilidad y asegurar la convergencia de los resultados.
- **ns0**: Es el número de modelos iniciales generados aleatoriamente en la primera etapa de la inversión. Este valor permite que el algoritmo explore un amplio rango de soluciones posibles antes de empezar el refinamiento.
- **ns**: Representa el número de modelos evaluados en cada iteración del algoritmo de inversión. Este parámetro es importante para equilibrar la precisión de la búsqueda y el tiempo de cálculo requerido.
- **nr**: Es el número de modelos seleccionados como los mejores en cada iteración para guiar la generación de modelos en la siguiente iteración. Un valor adecuado de nr ayuda a evitar que el algoritmo se quede atrapado en mínimos locales.
- **nmodels**: Este término generalmente representa el número total de modelos evaluados en una inversión. Es un producto de ns y el número de iteraciones, proporcionando una estimación del tamaño de la búsqueda total.
- **nrayleigh**: Indica el número de modos de onda de Rayleigh que se incluyen en el cálculo de la inversión. La inclusión de múltiples modos de Rayleigh puede ayudar a mejorar la precisión de las estimaciones de velocidad de onda de corte en la inversión.
- **nlove**: Similar a nrayleigh, este parámetro se refiere al número de modos de onda Love que se consideran en la inversión. Los modos Love son útiles para caracterizar perfiles de velocidad en suelos y rocas estratificados.
- **dcfmin**: Representan los valores mínimo y máximo de la frecuencia crítica (critical frequency) que se utilizarán en la inversión. Esta frecuencia delimita el rango de frecuencias sobre el cual se ajustarán las curvas de dispersión.
- **dcfmax**: Representan los valores mínimo y máximo de la frecuencia crítica (critical frequency) que se utilizarán en la inversión. Esta frecuencia delimita el rango de frecuencias sobre el cual se ajustarán las curvas de dispersión.
- **dcfnum**: Es el número de frecuencias críticas dentro del rango especificado por dcfmin y dcfmax. Este parámetro controla la resolución espectral de la curva de dispersión modelada en el proceso de inversión.
- **nellipticity**: Define el número de modos de elipticidad incluidos en la inversión. La elipticidad de las ondas de Rayleigh se puede utilizar para ajustar el modelo de velocidad, especialmente en perfiles de subsuelo complejos.
- **ellfmin**: Son los valores mínimo y máximo de la frecuencia para el análisis de elipticidad. Estos valores limitan el rango de frecuencias en el que se calcula la elipticidad de las ondas de Rayleigh.
- **ellfmax**: Son los valores mínimo y máximo de la frecuencia para el análisis de elipticidad. Estos valores limitan el rango de frecuencias en el que se calcula la elipticidad de las ondas de Rayleigh.
- **ellfnum**: Es el número de frecuencias en el rango definido por ellfmin y ellfmax. Controla la cantidad de puntos en la curva de elipticidad, influenciando la resolución de esta curva en el modelo final de inversión.

### Recomendaciones: 

- **ntrial**: Se recomienda realizar al menos **10** pruebas independientes para asegurar la convergencia y confiabilidad de los resultados

- **ns0**: Un valor recomendado es **10,000** modelos iniciales. Aunque ns0 tiene poco efecto en la convergencia, un número alto asegura que se evalúe un buen conjunto de modelos iniciales antes de la primera iteración del algoritmo

- **ns**: El número de modelos por iteración. Se sugiere un valor de **200** modelos por iteración en combinación con las iteraciones necesarias para alcanzar el número total de modelos buscado

- **nr**: Se recomienda un valor de alrededor de **100** para equilibrar entre convergencia rápida y evitar producir valores atípicos. Si persisten valores atípicos, se puede incrementar nr para hacer la inversión más exploratoria, aunque también requerirá aumentar el número total de modelos (It * Ns)

- **nmodels (It * Ns)**: Un número total de **50,000** modelos es sugerido como un compromiso para una búsqueda exhaustiva sin incurrir en un alto costo computacional

- **nrayleigh y nlove**: Aunque el documento no especifica un valor exacto, recomienda considerar múltiples modos de ondas de Rayleigh y Love según sea necesario para mejorar la precisión en los perfiles de velocidad de corte.

- **dcfmin y dcfmax**: Estos representan los límites inferior y superior de la frecuencia crítica en Hz para el cálculo de dispersión. La elección depende del rango de frecuencia de interés en el sitio específico y la profundidad de investigación deseada.

- **dcfnum**: Se debe seleccionar un número suficiente de frecuencias dentro del rango definido por dcfmin y dcfmax para asegurar una resolución adecuada en la curva de dispersión, pero el valor óptimo dependerá de la especificidad del sitio y las condiciones del experimento.

- **nellipticity**: El documento no proporciona valores específicos, pero se recomienda establecer el número de puntos de frecuencia suficientes para capturar la elipticidad de las ondas de Rayleigh con detalle en el rango de frecuencia de interés.

- **ellfmin y ellfmax**: Representan los límites de frecuencia para el análisis de elipticidad. Su valor depende del rango de frecuencia donde se espera que la elipticidad tenga mayor impacto en el análisis del perfil de velocidad de ondas de corte.

```python
# Minimum and maximum for all parameters. Refer to detailed instructions above.
vp_min, vp_max, vp_dec = 100., 5000., False
vs_min, vs_max, vs_dec = 80., 3500., False
pr_min, pr_max = 0.2, 0.5
rh_min, rh_max = 2000., 2000.

# Layering by Number (LN) parameterizations to consider. Add or remove as desired.
# See Vantassel and Cox (2021) for details.
lns = [3, 4, 5]

# Layering Ratios (LRs) parameterizations to consider. Add or remove as desired.
# See Vantassel and Cox (2021) and Cox and Teague (2016) for details.
lrs = [3.0, 2.0, 1.5]

# Depth factor, typically 2 or 3.
depth_factor = 2

# Minimum and maximum wavelength, selected from experimental disperison data by default.
wmin, wmax = min(target.wavelength), max(target.wavelength)


# Mass density.
if (rh_min - rh_max) < 1:
    rh = swprepost.Parameter.from_fx(rh_min)
else:
    rh = swprepost.Parameter.from_ln(wmin=wmin, wmax=wmax, nlayers=1, par_min=rh_min, par_max=rh_max, par_rev=False)
    
# Poisson's ratio
if (pr_max - pr_min) < 0.05:
    raise ValueError(f"Difference between pr_min and pr_max is too small ({pr_max-pr_min:2f}<0.05), use larger range.")
else:
    pr = swprepost.Parameter.from_ln(wmin=wmin, wmax=wmax, nlayers=1, par_min=pr_min, par_max=pr_max, par_rev=False)

# Make 1_parameters directory.
if not os.path.isdir("1_parameters/"):
    os.mkdir("1_parameters/")

# Parameterize Vs using Layering by Number (LN)
for ln in lns:
    vs = swprepost.Parameter.from_ln(wmin=wmin, wmax=wmax, nlayers=ln, par_min=vs_min, par_max=vs_max, par_rev=vs_dec, depth_factor=depth_factor)
    vp = swprepost.Parameter.from_parameter_and_link(par_min=vp_min, par_max=vp_max, par_rev=vp_dec, existing_parameter=vs, ptype="vs")
    par = swprepost.Parameterization(vp=vp, pr=pr, vs=vs, rh=rh)
    par.to_param(f"1_parameters/ln{ln}", version=version)

# Parameterize Vs using Layering Ratio (LR)
for lr in lrs:
    vs = swprepost.Parameter.from_lr(wmin=wmin, wmax=wmax, lr=lr, par_min=vs_min, par_max=vs_max, par_rev=vs_dec, depth_factor=depth_factor)
    vp = swprepost.Parameter.from_parameter_and_link(par_min=vp_min, par_max=vp_max, par_rev=vp_dec, existing_parameter=vs, ptype="vs")
    par = swprepost.Parameterization(vp=vp, pr=pr, vs=vs, rh=rh)
    par.to_param(f"1_parameters/lr{int(lr*10)}", version=version)

nparam = len(lns) + len(lrs)
if len(glob.glob("1_parameters/*.param")) == nparam:
    print(f"All {nparam} .param files exist, you may proceed.")
```

### Trazado de dispersión

Un trazado de dispersión es una representación gráfica de la curva de dispersión que muestra cómo varía la velocidad de fase de las ondas superficiales (como las ondas Rayleigh o Love) con la frecuencia o la longitud de onda.

Función del trazado de dispersión:
- El trazado permite visualizar las propiedades elásticas del subsuelo, ya que diferentes frecuencias de las ondas se ven afectadas por diferentes profundidades.
- Es fundamental en el análisis sísmico, ya que sirve como referencia experimental para las inversiones. Los modelos teóricos generados en el proceso de inversión se ajustan a esta curva experimental para lograr un modelo preciso del perfil de velocidades del subsuelo.

Componentes del trazado de dispersión:
- Eje X: Representa la frecuencia o longitud de onda.
- Eje Y: Representa la velocidad de fase de las ondas superficiales.

```python
ndc = 1       # Number of dispersion curves to plot, may use "all".
nray = 1      # Number of Rayleigh-wave modes to plot, may use "all".
nlov = 1      # Number of Love-wave modes to plot, may use "all".

fig, axs = plt.subplots(ncols=2, sharey=True, figsize=(6,3), dpi=150)

# Plot the Theoretical Modes of Inversion Ground Models.
color_id = 0
for partype in dcs:
    for parnumber in dcs[partype]:
        best = bestseed[partype][parnumber]
        suite = dcs[partype][parnumber][best]
        label = f"{partype}={parnumber} {suite.misfit_repr(nmodels=ndc)}"
        
        color = colors[color_id % len(colors)]
        for dc_count, dcset in enumerate(suite):
            for mode in range(nray):
                try:
                    dc = dcset.rayleigh[mode]
                    axs[1].plot(dc.wavelength, dc.velocity, color=color, label=label, linewidth=0.7)
                    label=None
                    axs[0].plot(dc.frequency, dc.velocity, color=color, label=label, linewidth=0.7)
                except KeyError:
                    print(f"Could not find mode {mode}.")                    
            if dc_count+1 == ndc:
                break
        color_id += 1
    
# Plot the Experimental Dispersion Curve
ax = axs[0]
target.plot(ax=ax)

ax = axs[1]
target.plot(ax=ax, x="wavelength")
ax.legend(loc="center left", bbox_to_anchor=(1,0.5))
ax.set_ylabel("")

plt.show()
```

### Trazado VS

El trazado de Vs es un componente crítico en el proceso de inversión de ondas superficiales, ya que permite visualizar y evaluar la distribución de la velocidad de onda de corte en el subsuelo. La velocidad de onda de corte es fundamental para el análisis de respuesta sísmica y para caracterizar las propiedades mecánicas del suelo y roca.

Pasos y Recomendaciones para el Trazado de Vs

- **Misfit de Vs**: Se utiliza una métrica de ajuste o misfit para cuantificar qué tan bien se ajusta un perfil de Vs derivado de la inversión al modelo verdadero o al perfil objetivo. La función de misfit de Vs calcula el porcentaje de diferencia promedio entre el perfil de inversión y el perfil verdadero en múltiples profundidades. Un bajo valor de misfit indica un buen ajuste

- **Selección de Parámetros de Inversión**: Los valores de los parámetros como el número de capas y su grosor se determinan en función de la información específica del sitio (geología, registros de perforación). En ausencia de información detallada, el documento sugiere combinar parametrizaciones de "Layering by Number" (LN) y "Layering Ratio" (LR) para capturar la incertidumbre en el perfil de Vs

- **Análisis de Múltiples Perfiles**: Para reflejar la incertidumbre en el perfil de Vs, es recomendable analizar múltiples perfiles de inversión. El documento recomienda graficar los 100 perfiles de misfit más bajo para cada parametrización aceptada y calcular el desvío estándar logarítmico (σln,Vs) como una medida cuantitativa de la incertidumbre. Esto ayuda a identificar las partes del perfil donde hay mayor variabilidad e incertidumbre, típicamente en las fronteras de capa

- **Comparación de Parámetros de Inversión**: Para obtener resultados más confiables, se recomienda utilizar una variedad de parametrizaciones y rechazar aquellas que muestren un rendimiento significativamente inferior en términos de misfit. Esto permite ajustar mejor la inversión a la complejidad del subsuelo y mejora la representación de la incertidumbre

- **Interpretación de Perfiles Resultantes**: A través del trazado de Vs, es posible evaluar tendencias de velocidad con la profundidad. En general, los perfiles de inversión tienden a capturar las tendencias de aumento de Vs con la profundidad. Sin embargo, el ajuste suele disminuir en profundidad debido a la naturaleza indeterminada del problema inverso y la limitación de datos en frecuencias largas, que afecta la resolución en capas profundas

### Trazando la incertidumbre

La incertidumbre en el proceso de inversión de ondas superficiales es un factor clave debido a la naturaleza no única del problema inverso, lo que significa que múltiples modelos pueden ajustarse igualmente bien a los datos experimentales. Es fundamental capturar y representar la incertidumbre asociada a los perfiles de velocidad de onda de corte (Vs) para una interpretación confiable de las propiedades del subsuelo.

Procedimiento para el Trazado de la Incertidumbre en Vs

- **Selección de Múltiples Perfiles de Baja Misfit**: Para reflejar la variabilidad en el modelo de Vs, es recomendable seleccionar y graficar los 100 perfiles con el misfit más bajo de cada parametrización aceptada. Estos perfiles representan las soluciones más cercanas al modelo experimental, permitiendo un análisis detallado de la variabilidad y la incertidumbre en Vs

- **Cálculo de la Desviación Estándar Logarítmica (σln,Vs)**: Para cuantificar la incertidumbre, el documento sugiere calcular la desviación estándar logarítmica de Vs (σln,Vs) entre los perfiles seleccionados. Este valor proporciona una medida cuantitativa de la dispersión en las velocidades de corte y permite identificar las zonas de mayor y menor incertidumbre dentro del perfil. La desviación estándar logarítmica suele ser más alta en áreas de transición de capas y en la base de la sección investigada, donde los datos de alta frecuencia tienen menos influencia

- **Representación Gráfica de la Incertidumbre**: Los perfiles de Vs de baja misfit se trazan con líneas delgadas o de tonos más claros, y se sobrepone una banda de incertidumbre que representa el rango de variabilidad de los perfiles en cada profundidad. Esto ayuda a visualizar las regiones de alta y baja confiabilidad en el modelo final

- **Interpretación del Rango de Incertidumbre**: El análisis visual de la banda de incertidumbre en el perfil de Vs permite identificar con precisión las secciones donde la inversión es más robusta y las zonas donde existe mayor ambigüedad en la estimación de Vs. En zonas profundas, la incertidumbre tiende a aumentar debido a la disminución en la resolución de las ondas superficiales para capas más profundas
